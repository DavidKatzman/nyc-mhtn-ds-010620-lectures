```python
# Load in libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.linear_model import LinearRegression
```

```python
# Create the List of Teams and convert to array
Teams_list = ['HOU', 'LAD', 'OAK', 'NYY', 'MIN', 'TBR', 
              'WSN', 'BOS', 'CLE', 'STL', 'CHC', 'NYM', 
              'ATL', 'ARI', 'TEX', 'MIL', 'CIN', 'PHI', 
              'LAA', 'CHW', 'TOR', 'SEA', 'COL', 'SFG', 
              'SDP', 'BAL', 'KCR', 'PIT', 'MIA', 'DET']
Teams_array = np.array(Teams_list)
```

```python
# Create the Total WAR list and convert to array
WAR_list = [67.6, 54.2, 50.3, 49.8, 49.4, 48.6, 45.5, 43.4, 41.5, 40.6, 40.3, 38.1, 38.1, 36.8, 33.3, 33.3, 29.3, 27.5, 26.4, 25.2, 25.1, 23.0, 22.8, 20.7, 18.5, 17.2, 16.6, 15.2, 10.1, 8.9]
WAR_array = np.array(WAR_list)
```

```python
# Determine Total WAR for the entire MLB (due to rounding number is slightly off what it should be)
WAR_array.sum()
```

```python
# Create the First Order Win list and convert to array
first_order_win_list = [108.1, 108.2, 97.5, 100.0, 98.2, 92.9, 95.5, 87.7, 92.8, 91.8, 90.7, 86.4, 91.9, 87.9, 74.5, 81.3, 79.9, 79.0, 71.5, 68.2, 70.9, 68.3, 70.0, 71.1, 70.0, 58.0, 63.5, 66.6, 60.8, 47.9]
first_order_win_array = np.array(first_order_win_list)
```

```python
# Determine Total First Order Wins for the entire MLB (due to rounding number is slightly off what it should be)
first_order_win_array.sum()
```

```python
# Create the Actual Wins List and convert to array
Actual_wins_list = [107, 106, 97, 103, 101, 96, 93, 84, 93, 91, 84, 86, 97, 85, 78, 89, 75, 81, 72, 72, 67, 68, 71, 77, 70, 54, 59, 69, 57, 47]
Actual_wins_array = np.array(Actual_wins_list)
```

```python
# Determine Total Wins for the entire MLB (because of the cancelled Tigers Cubs game, there were only 2429 total wins)
Actual_wins_array.sum()
```

```python
# Create dictionary of teams, WAR, First Order Wins, and Actual Wins
Team_Dict = {'Team': Teams_array, 'Total WAR': WAR_array, 'First Order Wins': first_order_win_array, 'Actual Wins': Actual_wins_array}
```

```python
# Convert dictionary into a DataFrame
Team_Wins_df = pd.DataFrame(Team_Dict)
```

```python
# Show DataFrame
Team_Wins_df
```

```python
# Prepare Variables for Linear Regression on Actual Wins
x = np.array(Team_Wins_df['Total WAR']).reshape((-1, 1))
y = np.array(Team_Wins_df['Actual Wins'])
model = LinearRegression()
```

```python
# Fit Linear Regression
model.fit(x,y)
```

```python
# Calculate R-Squared value
r_sq = model.score(x, y)
print('R-Squared value:', r_sq)
```

```python
# Calculate intercept
intercept = model.intercept_
intercept
```

```python
# Calculate slope
slope = model.coef_
slope
```

```python
# Set x and y values for regression line
x_values = Team_Wins_df['Total WAR']
y_values = intercept + slope * x_values
```

```python
# Plot Regression Line and actual data with labels for Actual Wins
Team_Wins_df.plot.scatter(x = 'Total WAR', y = 'Actual Wins', figsize = (20, 20))
for i in range(len(Team_Wins_df['Total WAR'])):
    plt.text(x[i]+.1, y[i]+.1, Team_Wins_df['Team'][i])
plt.plot(x_values, y_values)
plt.title('Actual Wins vs. Total Team WAR')
plt.xlabel('Total Team WAR')
plt.ylabel('Actual Team Wins')
plt.show()
```
